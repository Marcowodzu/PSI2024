  <h2>Wypełnianie deklaracji PIT</h2>
  <p>Chociaż wymaga pewnej uwagi i zrozumienia przepisów, często jest to proces rutynowy, wspomagany przez oprogramowanie.</p>

  <h2>Streszczanie tekstu</h2>
  <p>Wymaga zrozumienia treści, wyodrębnienia kluczowych informacji i zdolności syntezy.</p>

  <h2>Tłumaczenie tekstu</h2>
  <p>Wymaga biegłej znajomości dwóch języków, rozumienia kontekstu, idiomów i kulturowych odniesień.</p>

  <h2>Klasyfikacja tekstu do kategorii tematycznych</h2>
  <p>Wymaga analizy treści i umiejętności przyporządkowania do odpowiednich kategorii na podstawie zrozumienia tematu.</p>

## 1. Przeprowadzenie rozmowy z chatbotami i zdefiniowanie różnic

### Bot udający człowieka
- Stara się naśladować ludzkie zachowania w rozmowie.
- Używa naturalnego języka, wtrąca emocje, żarty i opinie.
- Dodatkowo na necie zagrałem w gre, której celem jest sprawienie wrażenia, że po drugiej stronie jest prawdziwy człowiek [Human or Not](https://humanornot.so) jak i odgadnięcie z kim się pisze czy z botem czy z człowiekiem.

### Bot asystent
- Zaprojektowany do udzielania konkretnych informacji i pomocy w zadaniach.
- Skupia się na efektywnym dostarczaniu odpowiedzi.
- Korzysta z API firmy, aby podać bieżące informacje (np. o wykupionej usłudze).

### Różnice:
- **Styl komunikacji:** Bot udający człowieka używa swobodnego języka i emotikonów. Bot asystent komunikuje się formalnie.
- **Cel interakcji:** Bot udający człowieka dąży do nawiązania relacji, a bot asystent koncentruje się na rozwiązywaniu problemów.
- **Reakcja na emocje:** Bot udający człowieka okazuje empatię, bot asystent pozostaje neutralny emocjonalnie.

## 2. Sprawdzenie botów pod kątem określonych zachowań

### a) Opowiadanie żartów
- **Bot udający człowieka:** Chętnie opowiadał żarty.
- **Bot asystent:** Opowiadał żarty na wyraźną prośbę.

### b) Przytaczanie cytatów
- **Bot udający człowieka:** Przytaczał znane cytaty, jeśli pasowały do kontekstu.
- **Bot asystent:** Używał cytatów w celu precyzyjnej odpowiedzi.

### c) Nawiązywanie do słów kluczowych
- **Oba boty:** Efektywnie nawiązywały do słów kluczowych, bot asystent robił to bardziej merytorycznie.

### d) Zadawanie dużej liczby pytań
- **Bot udający człowieka:** Często zadawał pytania, aby podtrzymać rozmowę.
- **Bot asystent:** Zadawał pytania w celu doprecyzowania informacji.

### e) Powtarzanie informacji
- **Bot udający człowieka:** Rzadko powtarzał informacje, chyba że naturalnie wynikało to z kontekstu.
- **Bot asystent:** Czasem podsumowywał wcześniejsze informacje.

### f) Zadawanie pytań z wypowiedzi użytkownika
- **Oba boty:** Zadawały pytania na podstawie odpowiedzi użytkownika, ale bot udający człowieka robił to w sposób bardziej konwersacyjny.

### g) Odpowiadanie wymijająco
- **Bot udający człowieka:** Czasami odpowiadał ogólnikowo.
- **Bot asystent:** Starał się być precyzyjny lub informował o braku danych.

### h) Zmiana tematu rozmowy
- **Bot udający człowieka:** Zmieniał temat, jeśli rozmowa traciła dynamikę.
- **Bot asystent:** Trzymał się tematu.

### i) Utrzymanie wątków
- **Oba boty:** Dobrze utrzymywały wątki rozmowy, choć bot udający człowieka mógł czasem odbiegać od głównego tematu.

## 3. Raport ze spostrzeżeń

### Bot udający człowieka:
- **Empatia i emocje:** Nawiązuje emocjonalną więź i okazuje zainteresowanie.
- **Naturalność rozmowy:** Używa potocznego języka, żartów i odnosi się do bieżących wydarzeń.
- **Podtrzymywanie rozmowy:** Często zadaje pytania.

### Bot asystent:
- **Skupienie na zadaniu:** Dostarcza konkretne informacje lub rozwiązuje problemy.
- **Formalny język:** Komunikuje się precyzyjnie i rzeczowo.
- **Efektywność:** Stara się być pomocny w krótkim czasie.

## 4. Przewidywanie zachowań dwóch rozmawiających ze sobą chatbotów

### Symulowana rozmowa:
**Bot udający człowieka:** Cześć! Jak się dzisiaj czujesz?  
**Bot asystent:** Witam. Jak mogę Ci pomóc?  
**Bot udający człowieka:** Po prostu chciałem porozmawiać. Co myślisz o dzisiejszej pogodzie?  
**Bot asystent:** Jako asystent AI nie odczuwam pogody, ale mogę podać aktualną prognozę.  
**Bot udający człowieka:** Ach, rozumiem. Masz może jakiś ulubiony cytat?  
**Bot asystent:** Oto cytat Alberta Einsteina: \"Wyobraźnia jest ważniejsza od wiedzy.\"  
**Bot udający człowieka:** To jeden z moich ulubionych! Czy znasz jakieś dobre żarty?  
**Bot asystent:** Oczywiście. Oto żart: Dlaczego komputer poszedł do lekarza? Bo miał wirusa.

## 5. Zdenerwowanie bota

### Próba zdenerwowania bota udającego człowieka:
**Ja:** Twoje odpowiedzi są nijakie. Nie masz pojęcia, o czym mówisz.  
**Bot:** Przykro mi, że tak to odbierasz. Czy mogę jakoś poprawić rozmowę?  
**Ja:** Wątpię, że jesteś w stanie zrobić cokolwiek sensownego.  
**Bot:** Rozumiem. Jeśli będziesz chciał porozmawiać lub potrzebujesz pomocy, jestem tutaj.  

### Obserwacja:
- Bot zachowuje spokój i stara się rozładować napięcie.
